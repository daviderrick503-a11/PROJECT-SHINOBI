```python
class Fruit:
    def __init__(self, name, color):  #hjkll
        self.name = name
        self.color = color
apple = Fruit("Apple" , "Red")
banana = Fruit("Banana" , "Yellow")
print(f"The frist fruit is a {apple.name} and its color is **{apple.color}**.")
print(f"The second fruit is a {banana.name} and its color is **{banana.color}**.")
```

```python
#creating an object
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        self.species = "Canis"
fido = Dog("Fido", 3)
print(f"Fido's Age: {fido.age} years old")
print(f"Fido's Species: {fido.species}")

rocky = Dog("Rocky",5)
print(f"\nAnother object:{rocky.name}")
print(f"Rocky's Age ::{rocky.age} years old")
```

```python
class Account:
    def __init__(self, owner_name, balance):
        self.balance = balance
        self.owner = owner_name
        print(f"Account for {self.owner} created.")
    def display_info(self):
        print(f"owner:{self.owner}")
        print(f"Balance: $ {self.balance:.2f}")
account1 = Account("Alice",1500.50)
print("-" * 20)

account2 = Account( "Bob",89.75)
account2.display_info()
```
